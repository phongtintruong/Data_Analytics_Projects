**Nấm được chia làm 2 loại:**
* Nấm độc: Nấm độc là những loại nấm không ăn được do chứa nhiều hàm lượng độc tố bên trong, gây ra bệnh tiêu hóa, dị ứng, ảo giác… Nấm độc thường có màu sắc sặc sỡ.
* Nấm ăn: Nấm ăn được sử dụng rộng rãi làm thực phẩm, sử dụng trong nhiều món ăn, ở nhiều nền ẩm thực khác nhau. Nấm là thực phẩm có giá trị dinh dưỡng cao, có độ đạm cao, đầy đủ các acid amin thiết yếu, ít chất béo và là những acid béo chưa bão hòa, tốt cho sức khỏe. Bên cạnh đó nấm còn có giá trị năng lượng cao, giàu khoáng chất và các vitamin, chứa nhiều vitamin nhóm B và C. Dù nấm không phải là nguồn vitamin D đáng kể, nhưng hàm lượng vitamin D có thể tăng lên khi được phơi với ánh sáng (nhất là tia cực tím) dù điều này làm thẫm lớp vỏ của chúng. Nấm cũng chứa nhiều nguyên tố vi lượng như: Sắt, Selen, Natri, Kali, Magie và Photpho. Trong nấm còn chứa nhiều hoạt chất có tính sinh học nên góp phần ngăn ngừa và điệu trị bệnh cho con người, hầu như các loài nấm ăn đều có tác dụng phòng ngừa chống u bướu, có dược tính rất mạnh với các căn bệnh nan y hiện nay như viêm gan, ung thư, HIV v.v
