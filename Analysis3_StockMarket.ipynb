```python
returns = pd.DataFrame()
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
for tick in tickers:
    returns[tick+' Return'] = df[tick]['Close'].pct_change()
returns
# dataframe returns chứa dữ liệu phần trăm thay đổi của giá qua từng ngày_
```

```python
#returns.min()
returns.idxmin()
# Ta thấy có 4 ngân hàng vào ngày 20/1/2009 có phần trăm tụt giá lớn nhất
```

```python
returns.idxmax()
# Phần lớn các ngân hàng có tỉ lệ tăng trưởng khá tốt vào nửa cuối năm 2008
```

```python
returns.describe()
#lợi nhuận trung bình của Citi bank là thấp nhất
#Chọn các cổ phiếu có std thấp để giảm thiểu rủi ro
```

# 2.2) Phân tích và kết luận

# 2.2.1) BAC

```python
df_BAC = df['BAC'][['Open', 'High', 'Low', 'Close','Volume']]
```

```python
qf = cf.QuantFig(df_BAC,legend='top',name='BAC')
qf.add_sma([144],width=1,color=['red'])
qf.add_volume()
qf.iplot()
# SMA là đường trung bình giá của 144 ngày gần nhất = (n1+n2 + ... + n144)/144
# Biểu đồ sử dụng là biểu đồ nến(candle stick) phổ biến trong tài chính
# QuantFig là class thuộc thư viện cufflinks
```

# Nhận xét: 
1. Có thể thấy đường SMA(144) có thể dùng lọc giá cổ phiếu tốt. Trong giai đoạn tăng, giá hầu như nằm trên đường SMA.
2. Phân tích 3 giai đoạn tháng 2/2009-4/2010, 1/2012-4/2014, 3/2016-now
3. Ở 2 giai đoạn đầu ta thấy khối lượng giao dịch lớn ở khi bắt đầu tăng và giai đoạn giữa. Càng về cuối sóng tăng thì khối lượng giao dịch càng giảm. 

```python
df_BAC_2016 = df['BAC']['2016':][['Open', 'High', 'Low', 'Close','Volume']]
```

```python
qf = cf.QuantFig(df_BAC_2016,legend='top',name='BAC')
qf.add_sma([144],width=1,color=['red'])
qf.add_volume()
qf.iplot()
```

# Nhận xét:
Áp dụng tính chất trên ta phân tích giá giai đoạn 3/2016- now
1. Có thể thấy từ bắt đầu đến giữa khoảng thời gian này, khối lượng giao dịch tương đối lớn so với tháng 7/2017 trở đi
